```python
from requests import get
url = 'https://www.flipkart.com/search?q=smart+phones&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_5_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_2_5_na_na_na&as-pos=2&as-type=RECENT&suggestionId=smart+phones&requestId=5a108fc5-586f-41d4-8684-8ab508cf5888&as-backfill=on&page=1'
response = get(url)
#getting web document that is html code
print(response.text[:500])
```

```python
from bs4 import BeautifulSoup as soup
page=soup(response.text,"html.parser")
container=page.find_all('div',class_='_1-2Iqu row')
```

```python
 print(soup.prettify(container[0]))
    #prettify makes your html code in structured format
```

```python
names=[]
ratings=[]
prices=[]
```

```python
import re
for co in container:
    
    Name=co.findAll("div",{"class":"_3wU53n"})

    Name=re.sub('^[^a-z,A-Z]'," ",str(Name).strip(']') )
    Name=re.sub("[,<\(\[].*?[\)>]", "", str(Name))
    rating=co.findAll("div",{"class":"hGSR34"})[0].text.strip()
    #name=co.findAll("div",{"class":"3wU53n"})[0].text.strip()
    price=co.findAll("div",{"class":"_1vC4OE _2rQ-NK"})[0].text.strip()
    
    names.append(Name)
    prices.append(price)
    ratings.append(rating)
```

```python
import pandas as pd

data= pd.DataFrame({'name': names,
'rating': ratings,
'price': prices,

})
print(data.info())
data
```
